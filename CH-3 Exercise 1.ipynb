```python
import numpy as np
import matplotlib.pyplot as plt

# Values for n
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

# Loop over different values of n
for n in n_values:
    # Step 2: Simulate throwing two dice n times
    dice1 = np.random.randint(1, 7, n)  # Random values between 1 and 6 for the first die
    dice2 = np.random.randint(1, 7, n)  # Random values between 1 and 6 for the second die
    s = dice1 + dice2  # Sum of the two dice

    # Step 3: Compute the histogram
    h, h2 = np.histogram(s, bins=np.arange(2, 14))  # Range from 2 to 12 (inclusive)

    # Step 4: Plot the histogram using a bar chart
    plt.figure(figsize=(8, 5))
    plt.bar(h2[:-1], h/n, width=0.8, align='edge', edgecolor='black')  # Normalize by n
    plt.title(f"Histogram of Dice Rolls (n = {n})")
    plt.xlabel('Sum of Dice')
    plt.ylabel('Probability')
    plt.xticks(np.arange(2, 13, 1))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
